```python
import cv2
import matplotlib.pyplot as plt


image_path = '/content/sample_data/Cat.jpg'
image = cv2.imread(image_path)


image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Display the image using Matplotlib
plt.imshow(image_rgb)
plt.axis('off')  # Turn off axis numbers and ticks
plt.show()
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
new_width = 300

# Calculate the new height while maintaining the aspect ratio
aspect_ratio = image.shape[1] / image.shape[0]
new_height = int(new_width / aspect_ratio)

# Resize the image
resized_image = cv2.resize(image, (new_width, new_height))

# Save the resized image
cv2.imwrite('resized_image.jpg', resized_image, [cv2.IMWRITE_JPEG_QUALITY, 90])

plt.imshow(cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB))
plt.axis('off')  # Turn off axis labels and ticks
plt.show()
```

```python
resized_image_path = 'resized_image.jpg'
cv2.imwrite(resized_image_path, resized_image, [cv2.IMWRITE_JPEG_QUALITY, 90])
print(f"Resized image saved as '{resized_image_path}' in the same folder as the original image.")
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply histogram equalization
equalized_image = cv2.equalizeHist(gray_image)

# Save the equalized image
equalized_image_path = 'equalized_image.jpg'
cv2.imwrite(equalized_image_path, equalized_image)
print(f"Equalized image saved as '{equalized_image_path}'.")

plt.figure(figsize=(10, 5))

plt.subplot(121)
plt.imshow(image, cmap='gray')
plt.title('Original Image')

plt.subplot(122)
plt.imshow(equalized_image, cmap='gray')
plt.title('Equalized Image')

plt.tight_layout()
plt.show()
```

```python
brightness_increase = 50
brightened_image = np.clip(image + brightness_increase, 0, 255).astype(np.uint8)

# Save the brightened image
brightened_image_path = 'brightened_image.jpg'
cv2.imwrite(brightened_image_path, brightened_image)
print(f"Brightened image saved as '{brightened_image_path}'.")

plt.imshow(cv2.cvtColor(brightened_image, cv2.COLOR_BGR2RGB))
plt.axis('off')  # Turn off axis
plt.show()
```

```python
# Fliping Vertically
flipped_image = cv2.flip(image, 1)

# Convert the image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Crop the image
x, y, width, height = 100, 100, 300, 300
cropped_image = image[y:y+height, x:x+width]

# Display the flipped, grayscale, and cropped images using Matplotlib
plt.figure(figsize=(10, 10))

plt.subplot(131)
plt.imshow(cv2.cvtColor(flipped_image, cv2.COLOR_BGR2RGB))
plt.title('Flipped Image')

plt.subplot(132)
plt.imshow(gray_image, cmap='gray')
plt.title('Grayscale Image')

plt.subplot(133)
plt.imshow(cv2.cvtColor(cropped_image, cv2.COLOR_BGR2RGB))
plt.title('Cropped Image')

plt.tight_layout()
plt.show()
```
